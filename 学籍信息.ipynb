校区号根据专业码确定，0~6999为1.0，7000多为2.0，980000多为3.0

```python
for i in df_xjjbsjxx.index:
    if df_xjjbsjxx.loc[i,'zym'] < 7000:
        df_xjjbsjxx.xqh = 1.0
    elif df_xjjbsjxx.loc[i,'zym'] < 8000:
        df_xjjbsjxx.xqh = 2.0
    elif df_xjjbsjxx.loc[i,'zym'] >= 8000:
        df_xjjbsjxx.xqh = 3.0
```

观察发现，学制信息丢失的学生，xslbm为5，6，或者7。xslbm为5或者7的同学学制均为1.0。xslbm为6的同学学制均为3.0

```python
for i in df_xjjbsjxx.index:
    if df_xjjbsjxx.loc[i,'xslbm'] == 7:
        df_xjjbsjxx.xz = 1.0
    elif df_xjjbsjxx.loc[i,'xslbm'] == 5:
        df_xjjbsjxx.xz = 1.0
    elif df_xjjbsjxx.loc[i,'xslbm'] == 6:
        df_xjjbsjxx.xz = 3.0
```
